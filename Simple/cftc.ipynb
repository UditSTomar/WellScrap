```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
from dateutil.parser import parse
from time import sleep, time
from random import randint
import re
```

```python
URL="https://cftc.gov/LawRegulation/DoddFrankAct/registerswapdealer.html"
r = requests.get(URL)
soup = BeautifulSoup(r.content, 'html5lib')
content=soup.find("div",attrs={'class':"region region-content"}
                 ).find('div',
                  attrs={'class':"field field--name-body field--type-text-with-summary field--label-hidden field--item"})
```

```python
all_companies=[]
for row in content.find_all('td'):
    all_companies.append(row.text.strip())
```
