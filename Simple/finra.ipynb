```python
URL="https://www.finra.org/about/firms-we-regulate/funding-portals-we-regulate"
r = requests.get(URL)
soup3 = BeautifulSoup(r.content, 'html5lib')
```

```python
content3=soup3.find_all("div",attrs={'class':"multicolumn-container"})
```

```python
all_companies_3=[]
for company_content in content3:
    company={}
    company["company_name"]=company_content.p.a.strong.text.strip()
    company["link"]=company_content.p.a.get("href")
    company_content.a.decompose()
    company["SEC_File_No"]=company_content.p.text.split("\n")[1].split(":")[1].strip()
    company["website"]=company_content.p.text.split("\n")[3][company_content.p.text.split("\n")[3].find("https"):].strip()
    company["refined_address"]=company_content.p.text.split("\n")[4].strip()
    add_info=company_content.p.text.split("\n")[5]
    if(",") in add_info:
        company["city"]=add_info.split(",")[0].strip()
        company["state"]=add_info.split(",")[1].strip()[0:2].strip()
        company["zip_code"]=add_info.split(",")[1].strip()[2:].strip()
    else:
        company["city"]=np.nan
        company["state"]=add_info[:add_info.find(" ")].strip()
        company["zip_code"]=add_info[add_info.find(" ")+1:].strip()
        
    all_companies_3.append(company)
```

```python
all_companies_3=pd.DataFrame(all_companies_3)
```
