```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
from dateutil.parser import parse
from time import sleep, time
from random import randint
import re
import numpy as np
```

```python
URL="https://www.finra.org/about/firms-we-regulate/broker-dealer-firms-we-regulate"
r = requests.get(URL)
soup1 = BeautifulSoup(r.content, 'html5lib')
```

```python
content1=soup1.find("div",attrs={'id':'main-wrapper'}).find("div",attrs={'class':"views-field views-field-body"})
```

```python
all_companies_1=[]
for company_content in (content1).find_all('p')[:len(content1.find_all('p'))-1]:
    company={}
    company_inf=company_content.contents
    company["company_name"]=company_inf[0]
    add_info=company_inf[2].split(",")
    company["refined_address"]=company_refined_address=",".join(add_info[0:-2]).strip()
    company["city"]=add_info[-2].strip()
    company["state"]=add_info[-1].strip()[0:2].strip()
    company["zip_code"]=add_info[-1].strip()[add_info[-1].strip().find(" ")+1:].strip()
    company["company_mailing_address"]=company_inf[4] if len(company_inf)==5 else np.nan
        
    all_companies_1.append(company)
```

```python
all_companies_1=pd.DataFrame(all_companies_1)
```

```python
URL="https://www.finra.org/about/firms-we-regulate/funding-portals-we-regulate"
r = requests.get(URL)
soup3 = BeautifulSoup(r.content, 'html5lib')
```

```python
content3=soup3.find_all("div",attrs={'class':"multicolumn-container"})
```

```python
all_companies_3=[]
for company_content in content3:
    company={}
    company["company_name"]=company_content.p.a.strong.text.strip()
    company["link"]=company_content.p.a.get("href")
    company_content.a.decompose()
    company["SEC_File_No"]=company_content.p.text.split("\n")[1].split(":")[1].strip()
    company["website"]=company_content.p.text.split("\n")[3][company_content.p.text.split("\n")[3].find("https"):].strip()
    company["refined_address"]=company_content.p.text.split("\n")[4].strip()
    add_info=company_content.p.text.split("\n")[5]
    if(",") in add_info:
        company["city"]=add_info.split(",")[0].strip()
        company["state"]=add_info.split(",")[1].strip()[0:2].strip()
        company["zip_code"]=add_info.split(",")[1].strip()[2:].strip()
    else:
        company["city"]=np.nan
        company["state"]=add_info[:add_info.find(" ")].strip()
        company["zip_code"]=add_info[add_info.find(" ")+1:].strip()
        
    all_companies_3.append(company)
```

```python
all_companies_3=pd.DataFrame(all_companies_3)
```
