```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
from dateutil.parser import parse
from time import sleep, time
from random import randint
import re
from IPython.core.display import clear_output
from warnings import warn
from time import sleep
from random import randint
import numpy as np
URL = "https://fraternalalliance.org/associate-members"
r = requests.get(URL)
soup = BeautifulSoup(r.content, 'html5lib')
```

```python
companies = soup.find_all('div', attrs = {'class':'associate-member'}) 
```

```python
all_companies=[]
for new_company in companies:    
    company={}
    company["company_name"]=new_company.h2.text.strip()
    company["company_refined_address"]=new_company.find('p',attrs = {'class':"address"}).contents[0].strip()
    company_loc=re.split('\t|\n',new_company.find('p',attrs = {'class':"address"}).text)[-1].split(",")
    company["company_city"]=company_loc[0]
    company["company_state"]=" ".join(company_loc[1].split(" ")[0:-1])
    company["company_pin"]=company_loc[1].split(" ")[-1]
    contact=new_company.find('p',attrs = {'class':"primary-contact"})
    company["contact_name"]=contact.contents[0].split(",")[0].strip()
    company["contact_designation"]=contact.contents[0].split(",")[1].strip()
    company["contact_email_id"]=contact.find('a').text if contact.find('a') is not None else np.nan# or contact.find('a').get('href')
    company["company_bio"]=new_company.find("p",attrs={'class':"bio"}).text.strip()
    
    all_companies.append(company)
```

```python
companies=pd.DataFrame(all_companies)
```
